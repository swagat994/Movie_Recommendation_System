```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movies = pd.read_csv('data/movies.csv')
credits = pd.read_csv('data/credits.csv')
```

```python
movies.head()
movies.shape
movies.info()
movies.describe()
```

```python
movies.isnull().sum()
sns.heatmap(movies.isnull(), cbar=False)
```

```python
movies.duplicated().sum()
```

```python
movies = movies.merge(credits, on='title')
```

```python
movies['genres'].head()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(movies['popularity'], bins=50)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(movies['vote_average'], bins=20)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=movies['runtime'])
plt.show()
```

```python
movies['original_language'].value_counts().head(10)
```

```python
numeric_cols = [
    'budget',
    'revenue',
    'runtime',
    'popularity',
    'vote_average',
    'vote_count'
]

corr = movies[numeric_cols].corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True)
plt.show()
```
